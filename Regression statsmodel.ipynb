```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
mpg = pd.read_csv('mpg.csv')
mpg.dropna(subset = ['horsepower'], inplace = True)
mpg.head()
```

```python
mpg.drop(mpg.columns[[2,7,8]], axis=1, inplace=True)
```

```python
mpg.head()
```

```python
mpg.isnull().any()
```

```python
mpg.corr()
```

```python
X = mpg[['horsepower', 'weight']]
Y = mpg['mpg']
```

```python
X = sm.add_constant(X)
reg_res = sm.OLS(Y, X).fit()
print(reg_res.summary())
```
